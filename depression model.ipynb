## preprocessing the data 




```python
import re
import nltk

nltk.download('stopwords')
#from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
ps = LancasterStemmer()

#all_stopwrds = stopwords.words('english')
#all_stopwrds.remove('not')
```

```python
corpus=[]

for i in range(0,72312):
  tweet = re.sub('[^a-z A-Z]',' ',dataset["Tweets"][i])
  tweet = tweet.lower() # convert text to lower case
  tweet = tweet.split()# Splits at space
  tweet = [ps.stem(word) for word in tweet] #if not word in set(all_stopwrds)
  tweet = ' '.join(tweet) # join elements of text with space
  corpus.append(tweet)
```

```python
corpus
```

# text vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=200)
```

```python
x = cv.fit_transform(corpus).toarray()
y = dataset[" Label"].values
```

```python
len(x[0])
```

```python
import pickle
bow_path = '/content/drive/MyDrive/GP/bow.pkl'
pickle.dump(cv, open(bow_path,'wb'))
```

# Data splitting

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

# Training model with Decision Tree learning algorithm

```python
accuracy = []

from  sklearn.tree import DecisionTreeClassifier
tlc = DecisionTreeClassifier()
tlc.fit(x_train,y_train)
```

```python
tree_training_accuracy=tlc.score(x_train,y_train)
print(tree_training_accuracy)
accuracy.append(tree_training_accuracy)
```

```python
tree_test_accuracy=tlc.score(x_test,y_test)
print(tree_test_accuracy)
accuracy.append(tree_test_accuracy)
```
