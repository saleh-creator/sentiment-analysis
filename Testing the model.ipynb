```python
import re
import nltk

#nltk.download('stopwords')
#from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

#all_stopwrds = stopwords.words('english')
#all_stopwrds.remove('not')
```

```python
corpus=[]

for i in range(0,100):
  tweet1 = re.sub('[^a-zA-Z]',' ',df1["tweet"][i])
  tweet1 = tweet1.lower() # convert text to lower case
  tweet1 = tweet1.split()# Splits at space
  tweet1 = [ps.stem(word) for word in tweet1] #
  tweet1= ' '.join(tweet1) # join elements of text with space
  corpus.append(tweet1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle
bpath = "/content/drive/MyDrive/GP/bow.pkl"
cv=pickle.load(open(bpath,'rb'))
```

```python
x_fresh = cv.transform(corpus).toarray()
```
